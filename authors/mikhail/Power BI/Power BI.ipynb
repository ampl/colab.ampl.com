```python
# Install dependencies 
! pip install powerbiclient                         
from powerbiclient import Report, models        # install the Report and models module from the powerbiclient package
from io import StringIO
from ipywidgets import interact
import requests
import pandas as pd
print('Import necessary libraries.')
```

```python
# Import the Device Code Login Authentication class to authenticate against Power BI and initiate the Microsoft device authentication
#from powerbiclient.authentication import DeviceCodeLoginAuthentication
#device_auth = DeviceCodeLoginAuthentication() 

from powerbiclient.authentication import InteractiveLoginAuthentication
interactive_auth = InteractiveLoginAuthentication() 
```

```python
# Get the group and report ID from PowerBI service
group_id="8f0dd820-1f47-4cf6-81fa-17f85e6a88e5"     # "YOU HAVE TO PUT HERE YOUR POWER BI GROUP ID OR WORKSPACE ID

report_id="17a9afef-262c-41d9-8123-2f70505513e9"    # YOU HAVE TO PUT HERE YOUR POWER BI REPORT ID

#  Pass the group_id, report_id and device_auth to the report and print the report
report = Report(group_id=group_id, report_id=report_id, auth = interactive_auth) #device_auth)
report.set_size(600, 1024)
# output.enable_custom_widget_manager()

report              # Show the power BI report with the downloaded data.
```

```python
# Get the names of all pages
# for page in report.get_pages(): 
#   print(page['name'])

pages = report.get_pages() # 
# display_name = pages[2]["displayName"]        # Get the displayName of page 3
#print(display_name)                             # display the displayName of page 3
page_name=pages[2]["name"]                      # Get the name of page 3
print(page_name)                                # display the name of page 3
# print(pages)                                  # display the names of all pages

visuals = report.visuals_on_page(page_name)
#visuals                                        # show information of all visuals   
visual_name = 'a91a19945528d85eb6c8'            # Assign the name of the visual element to the variable !!! 
# visual_name = visuals[0]["name"]              # Assign the name of the visual element by it's number to the variable
```

```python
# !!!  It need to activate the PowerBI page with the selected visual
exported_data = report.export_visual_data(page_name, visual_name, rows=1000, export_data_type=models.ExportDataType.SUMMARIZED.value) # Max 30 000 rows
exported_data=StringIO(exported_data)       
df=pd.read_csv(exported_data, sep=",")
display(df)
```
